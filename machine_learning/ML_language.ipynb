```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv("./data/langues/langues.csv")
print(df.shape)
print("les colonnes: ", df.columns)
df.columns = ["english", "francais"]
df.head(20)
```

```python
dfWaxtane = pd.read_csv("./data/langues/waxtane.csv")
print(dfWaxtane.shape)
print("les colonnes: ", dfWaxtane.columns)
print(dfWaxtane.columns)
dfWaxtane.head(20)
```

```python
#utilisation des methodes de scrapping pour recolter olus de donnees concernant le wolof
from utils import ScrapSite

scrapping = ScrapSite()
data = scrapping.scanSite("https://www.socolas-blog.com/parler-senegal-vocabulaire-wolof-diola-peul/", "tr")
scrapping.recupData(data)
```

```python
#Selection des donnees en wolof
dataWolof = dfWaxtane["sentence_wolof"]
dataWolof.head()
```

```python
#division de dataframe pour avoir un en anglais et un autre en fr

dataFR = df["francais"]
dataEN = df["english"]
```

```python
dataLabelFR = []
for size in range(len(dataFR)):
    dataLabelFR.append("francais")
```

```python
dataLabelEN = []
for size in range(len(dataEN)):
    dataLabelEN.append("anglais")
```

```python
dataLabelWolof = []
for size in range(len(dataWolof)):
    dataLabelWolof.append("wolof")
```

```python
datasFR = {
    "sentences": dataFR,
    "langues": dataLabelFR
}

datasEN = {
    "sentences": dataEN,
    "langues": dataLabelEN
}

datasWolof = {
    "sentences": dataWolof,
    "langues": dataLabelWolof
}
```

```python
# new datatframe with fr and en

dfFR = pd.DataFrame(datasFR)
dfEN = pd.DataFrame(datasEN)
dfWO = pd.DataFrame(datasWolof)  

df = pd.concat([dfFR, dfEN, dfWO], ignore_index=True)

print("La taille du dataframe: ", df.shape)
df.head(20)
```

```python
#save models
df.to_csv("./data/langues/languages.csv")
```

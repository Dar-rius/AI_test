```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv("./data/langues/langues.csv")
print(df.shape)
print("lees colonnes: ", df.columns)
df.columns = ["english", "francais"]
df.head(20)
```

```python
#division de dataframe pour avoir un en anglais et un autre en fr

dataFR = df["francais"]
dataEN = df["english"]
```

```python
dataLabelFR = []
for size in range(len(dataFR)):
    dataLabelFR.append("francais")
```

```python
dataLabelEN = []
for size in range(len(dataEN)):
    dataLabelEN.append("anglais")
```

```python
datasFR = {
    "sentences": dataFR,
    "langues": dataLabelFR
}

datasEN = {
    "sentences": dataEN,
    "langues": dataLabelEN
}
```

```python
# new datatframe with fr and en

dfFR = pd.DataFrame(datasFR)
dfEN = pd.DataFrame(datasEN)

df = pd.concat([dfFR, dfEN], ignore_index=True)

print("La taille du dataframe: ", df.shape)
df.head(20)
```
